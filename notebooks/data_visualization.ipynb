```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load Dataset
file_path = r"C:\Users\anura\Desktop\Project 4- Climate Change Modeling\data\raw\climate_nasa.csv"  
data = pd.read_csv(file_path)
```

```python
# Convert `date` to datetime format
data['date'] = pd.to_datetime(data['date'], errors='coerce')
```

```python
# 2. Bar Plot (Top Profiles by Likes)
top_profiles = data.groupby('profileName')['likesCount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_profiles.values, y=top_profiles.index, palette='viridis')
plt.title('Top Profiles by Total Likes')
plt.xlabel('Total Likes')
plt.ylabel('Profile Name')
plt.show()
```

```python
# 4. Histogram (Distribution of Likes)
plt.figure(figsize=(8, 6))
sns.histplot(data['likesCount'], bins=30, kde=True, color='blue')
plt.title('Distribution of Likes Count')
plt.xlabel('Likes Count')
plt.ylabel('Frequency')
plt.show()
```

```python

# 6. Box Plot (Outlier Analysis)
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='likesCount', color='blue')
plt.title('Box Plot of Likes Count')
plt.show()


```

```python
# 7. Correlation Heatmap
numeric_data = data[['likesCount', 'commentsCount']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(numeric_data, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 8. Pie Chart (Proportion of Top Profiles)
top_profiles_pie = top_profiles / top_profiles.sum()
plt.figure(figsize=(8, 8))
plt.pie(top_profiles_pie, labels=top_profiles_pie.index, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(top_profiles_pie)))
plt.title('Proportion of Top Profiles by Likes')
plt.show()
```
